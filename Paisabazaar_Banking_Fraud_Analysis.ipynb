#### Chart - 4
#### Relation between New Monthaly balance Category v/s Delay from Due Date

````python
# Chart - 4 visualization code
# Create bins for 'Delay_from_due_date'
bins = list(range(0, int(Data['Delay_from_due_date'].max()) + 5, 5)) # Create bins every 5 days
labels = [f'{i}-{i+4}' for i in bins[:-1]]
Data['Delay_from_due_date_bins'] = pd.cut(Data['Delay_from_due_date'], bins=bins, labels=labels, right=False)

# Group by the new bins and New_Monthly_Balance_Category
delay_bins_balance_counts = Data.groupby(['Delay_from_due_date_bins', 'New_Monthly_Balance_Category'], observed=False).size().unstack(fill_value=0)

# Visualize the results
delay_bins_balance_counts.plot(kind='bar', stacked=True, figsize=(7, 4), colormap='RdYlBu') # Changed colormap
plt.title('New Monthly Balance Category by Delay from Due Date Bins')
plt.xlabel('Delay from Due Date (days)')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.legend(title='New Monthly Balance Category')
plt.tight_layout()
plt.show()
````

#### Chart - 8

#### Distribution of New Monthly Balance Distribution by Credit Score

````python
# Chart - 8 visualization code
plt.figure(figsize=(6, 4))
sns.boxplot(data=Data, x='Credit_Score', y='New_Monthly_Balance', palette='viridis', order=['Poor', 'Standard', 'Good'])
plt.title('New Monthly Balance Distribution by Credit Score')
plt.xlabel('Credit Score')
plt.ylabel('New Monthly Balance')
plt.tight_layout()
plt.show()
````

````python
# Calculate IQR and outliers for New_Monthly_Balance by Credit_Score
for score in Data['Credit_Score'].unique():
    print(f"Analyzing New_Monthly_Balance for Credit Score: {score}")
    subset_data = Data[Data['Credit_Score'] == score]['New_Monthly_Balance']

    Q1 = subset_data.quantile(0.25)
    Q3 = subset_data.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = subset_data[(subset_data < lower_bound) | (subset_data > upper_bound)]

    print(f"  IQR: {IQR:.2f}")
    print(f"  Lower Bound for Outliers: {lower_bound:.2f}")
    print(f"  Upper Bound for Outliers: {upper_bound:.2f}")
    print(f"  Number of Outliers: {len(outliers)}")
    print("-" * 30)
````

##### 1. Why did you pick the specific chart?

Answer

A box plot is a suitable visualization to compare the distribution of a continuous variable (`New_Monthly_Balance`) across different categories of a categorical variable (`Credit_Score`). It effectively shows the median, quartiles, and spread of the new monthly balance for each credit score group and helps in identifying potential outliers.

##### 2. What is/are the insight(s) found from the chart?
